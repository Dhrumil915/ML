```python
# Converting datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
# Extract Year
date['date_year'] = date['date'].dt.year
```

```python
date.head()
```

```python
# Extract Months
date['date_months'] = date['date'].dt.month_name()
```

```python
date.head()
```

```python
# Extract Day
date['date_day'] = date['date'].dt.day
```

```python
date.head()
```

```python
# Extract day of week
date['date_week'] = date['date'].dt.day_of_week
```

```python
date.head()
```

```python
# Day of week name
date['date_week_name'] = date['date'].dt.day_name()
```

```python
date.head()
```

```python
# Check is weekend
date['date_is_weekend'] = np.where(date['date_week_name'].isin(['Sunday', 'Saturday']), 1,0)
```

```python
date.head()
```

```python
# Extract Quater
date['quarter'] = date['date'].dt.quarter
```

```python
date.head()
```

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)
```

```python
date.head()
```

```python
# Extract time elasped between dates
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Month Passed
np.round((today -date['date']) / np.timedelta64(1, 'D'),0)
```
